## ✅ Célula 1 — Python e localização do projeto

```python
import sys, os

v = sys.version_info
print(f'Python: {sys.version}')
print('✅ Versão compatível!' if v.major == 3 and v.minor >= 10 else '❌ Requer Python 3.10+')

def encontrar_raiz():
    caminho = os.path.abspath('.')
    marcadores = ('setup_projeto.ipynb', 'main_fase1.py', 'requirements.txt')
    for _ in range(6):
        if any(os.path.exists(os.path.join(caminho, m)) for m in marcadores):
            return caminho
        caminho = os.path.dirname(caminho)
    return os.path.abspath('.')

ROOT = encontrar_raiz()
os.chdir(ROOT)
if ROOT not in sys.path:
    sys.path.insert(0, ROOT)

print(f'\n📂 Raiz: {ROOT}')
print('   Todos os arquivos serão criados aqui.')
```

## 📁 Célula 2 — Pastas e `__init__.py`

```python
pastas = [
    'collectors', 'processors', 'models',
    'dashboard', 'dashboard/pages',
    'data/raw', 'data/processed', 'data/resultados',
    'notebooks', 'docs', 'logs', 'assets', '.github/workflows',
]
modulos = {'collectors', 'processors', 'models', 'dashboard', 'dashboard/pages'}

print('Criando pastas...\n')
for pasta in pastas:
    os.makedirs(pasta, exist_ok=True)
    open(os.path.join(pasta, '.keep'), 'w').close()
    init_info = ''
    if pasta in modulos:
        init_path = os.path.join(pasta, '__init__.py')
        if not os.path.exists(init_path):
            open(init_path, 'w').close()
            init_info = ' + __init__.py'
    print(f'  ✅ {pasta}/{init_info}')

print('\n✅ Estrutura criada!')
```
